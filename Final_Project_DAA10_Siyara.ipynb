**7. AQI Level Categorization**

• Creating categorical AQI levels using pd.cut() with standard air quality index ranges.

```python
df['AQI_level']= pd.cut(df['AQI'], bins=[0, 50, 100, 200, 500], labels=['Good','Moderate','Poor','Very Poor'])
print("AQI Level Distribution:")
print(df['AQI_level'].value_counts())
```

**8. Missing Values Treatment in AQI_level**

```python
mumbai_mode= df.loc[df['City']== 'Mumbai','AQI_level'].mode()[0]
df.loc[(df['City']=='Mumbai')& (df['AQI_level'].isnull()),'AQI_level']=mumbai_mode
```

**9. Outlier Detection for Pollutants**

• Used Interquartile Range (IQR) method to detect statistical outliers in all pollutant columns and AQI.

```python
pollutants = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3', 'AQI']

for col in pollutants:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"Number of outliers in {col}: {len(outliers)}")
```

**Output:**

• Count of outliers identified for each pollutant variable

• Provides insight into data distribution and extreme values

**Insight:**

```python

plt.figure(figsize=(8,5))
sns.boxplot(x='Season', y='AQI', hue='Season', data=df, palette='Set2',  width=0.6, linewidth=1.2)
plt.title('AQI Distribution Across Seasons', fontsize=14, weight='bold')

# 🔹 Add one horizontal reference line near the highest median (around Summer)
plt.axhline(y=df[df['Season']=='Winter']['AQI'].median(), color='red', linestyle='--', linewidth=1.2, 
label='Winter Median AQI')

# 🔹 Add a light grid for readability (optional)
plt.grid(axis='y', linestyle=':', alpha=0.4)

plt.legend(title='Reference', loc='upper right')
plt.tight_layout()
plt.show()
```

**Observation:**
The median AQI is slightly higher during Winter and slightly lower during Monsoon, though the variation across seasons is not very large.

**Insight:**
During winter, cooler temperatures and stagnant air trap pollutants near the ground, leading to higher AQI.
In contrast, monsoon rainfall helps wash out pollutants, improving air quality.
